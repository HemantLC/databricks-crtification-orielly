```sql
--Creating delta lake tables
CREATE TABLE product_info(
    product_id INT,
    product_name STRING,
    product_category STRING,
    product_price DOUBLE,
    product_quantity INT
)
USING DELTA;
```

```sql
-- Inserting Data into product_info table
INSERT INTO product_info VALUES
(1,'Laptop','Electronics',1000,10);

INSERT INTO product_info VALUES
(2,'Microwave','Kitchen', 249.5, 30),
(3,'Table','Furniture', 120, 20),
(4,'Monitor','Electronics', 300, 15);
```

```sql
SELECT * FROM product_info;
```

```sql
DESCRIBE DETAIL product_info;
```

```sql
-- Updating Delta lake tables
UPDATE product_info 
SET product_price = product_price + 100 
WHERE product_id =3; 
```

```sql
SELECT * from product_info;
```

```sql
--Exploring the table history

DESCRIBE HISTORY product_info;
```

```sql
-- Exploring delta-lake time travel
select * from product_info TIMESTAMP AS OF ('2026-06-17T19:11:41.000+00:00');
```

```sql
-- Quering by version number
Select * from product_info version as of 2;
```

```sql
select * from product_info@v3;
```

```sql
--Accidental delete
--DESCRIBE HISTORY product_info;
DELETE FROM product_info;
```

```sql
DESCRIBE HISTORY product_info;
```

```sql
-- Restore the product_info table
RESTORE TABLE product_info VERSION AS OF 3;
-- Verify the table has been restored
SELECT * FROM product_info;
```

```sql
DESCRIBE HISTORY product_info;
```
